## -10 products in different stores 
## -varying prices for the same product
## -duplicated entries for the same product in the same store

```python
import pandas as pd
from numpy.random import randint
df = pd.DataFrame(columns = ['host', 'store', 'upc', 'price'])
for i in range(30):
    df.loc[i] = ['host' + str(randint(1,5))] + ['store' + str(randint(1,10))] + [str(randint(1000,1010))] + list(randint(19,27, size=1))
df['price'] = pd.to_numeric(df['price'], downcast='float')
df.sort_values('store')
```

## -find mean price of products and assign to their corresponding UPCs
## -drop rows of duplicated entries of the same product in all stores

```python
mean_frame = df.groupby('upc')['price'].mean().to_frame().reset_index()
mean_frame
```

```python
df.loc[df.upc == '1000', ['upc', 'price']] = '1000', mean_frame.iloc[0,1]
df.loc[df.upc == '1001', ['upc', 'price']] = '1001', mean_frame.iloc[1,1]
df.loc[df.upc == '1002', ['upc', 'price']] = '1002', mean_frame.iloc[2,1]
df.loc[df.upc == '1003', ['upc', 'price']] = '1003', mean_frame.iloc[3,1]
df.loc[df.upc == '1004', ['upc', 'price']] = '1004', mean_frame.iloc[4,1]
df.loc[df.upc == '1005', ['upc', 'price']] = '1005', mean_frame.iloc[5,1]
df.loc[df.upc == '1006', ['upc', 'price']] = '1006', mean_frame.iloc[6,1]
df.loc[df.upc == '1007', ['upc', 'price']] = '1007', mean_frame.iloc[7,1]
df.loc[df.upc == '1008', ['upc', 'price']] = '1008', mean_frame.iloc[8,1]
df.loc[df.upc == '1009', ['upc', 'price']] = '1009', mean_frame.iloc[9,1]
                                                                                   
df.sort_values('store').drop_duplicates(subset=['store', 'upc', 'price'])
```

## -final output is a unique host-store-upc combination with a single price
